```python
import pandas as pd
import seaborn as sns
df = pd.read_csv('Clean_whiskey_data.csv')
df = df.drop('Unnamed: 0', axis = 1)
count = df.rating.size
print("Reviews : "+ str(count))
count = df.User_ID.unique().size
print("Unique Users count : "+ str(count))
count = df.Whiskey_ID.unique().size
print("Unique Whiskey_ID's : "+ str(count))
```

Dataframe : 2393 unique users, 2718 unique whiskies, and 67744 reviews.

```python
df.sample(n=3)
```

```python
ax_1 = sns.countplot(x="rating", data=df)
for ind, label in enumerate(ax_1.get_xticklabels()):
    if ind % 5 == 0:
        label.set_visible(True)
    else:
        label.set_visible(False)
ax_1
```

Most of the users rated the whiskies prety high ,reviews range from mid 85's to 100. Because most whiskies are part of the top rated whiskies, so they highly rated.

```python
ax_2 = sns.countplot(x="User_ID", data=df)
for ind, label in enumerate(ax_2.get_xticklabels()):
    if ind % 300 == 0:
        label.set_visible(True)
    else:
        label.set_visible(False)
ax_2
```

Most of users reviewed many different types of whiskies

```python
ax_6 = sns.displot(df.rating,kde=False)
```

```python
ax_3 = df.groupby('Whiskey_ID').count()['rating'].hist(bins = 500);
ax_3.set_xlim(1,100);
ax_3.set_xticks(range(1,100));
ax_3.set_xlabel('Number of Reviews')
ax_3.set_ylabel("Count of Whiskies_ID's")
ax_3.grid(False)

for ind, label in enumerate(ax_3.get_xticklabels()):
    if ind % 5 == 0:
        label.set_visible(True)
    else:
        label.set_visible(False)
ax_3
```

```python
ax_4 = df.groupby('Whiskey_ID').count()['User_ID'].hist(bins = 500);
ax_4.set_xlim(1,100);
ax_4.set_xticks(range(1,100));
ax_4.set_xlabel('Number of Reviews')
ax_4.set_ylabel("Count of user_ID's")
ax_4.grid(False)

for ind, label in enumerate(ax_4.get_xticklabels()):
    if ind % 5 == 0:  # every 10th label is kept
        label.set_visible(True)
    else:
        label.set_visible(False)

ax_4
```
