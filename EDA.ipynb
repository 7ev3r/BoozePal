```python
import pandas as pd
import seaborn as sns

df = pd.read_csv('Clean_whiskey_data.csv')
df = df.drop('Unnamed: 0', axis = 1)

count = df.User_ID.unique().size
print("Unique Users count : "+ str(count))
count = df.Whiskey_ID.unique().size
print("Unique Whiskey_ID's : "+ str(count))
```

Final dataframe has 1009 unique users, 463 unique whiskies, and 37567 total reviews.

```python
df.head()
```

```python
ax_1 = sns.countplot(x="rating", data=df)

for ind, label in enumerate(ax_1.get_xticklabels()):
    if ind % 2 == 0:
        label.set_visible(True)
    else:
        label.set_visible(False)
ax_1
```

Most of the users on rated the whiskies fairly highly as a majority of our reviews range from mid 80's to 100. These whiskies are part of the top 1000 rated whiskies, so they highly rated.

```python
ax_2 = sns.countplot(x="User_ID", data=df)
for ind, label in enumerate(ax_2.get_xticklabels()):
    if ind % 100 == 0:
        label.set_visible(True)
    else:
        label.set_visible(False)
ax_2
```

Majority of users having reviewed many different types of whiskies

```python
ax_3 = df.groupby('Whiskey_ID').count()['rating'].hist(bins = 300);
ax_3.set_xlim(1,50);
ax_3.set_xticks(range(1,150));
ax_3.set_xlabel('Number of Reviews')
ax_3.set_ylabel("Count of Whiskies_ID's")
ax_3.grid(False)

for ind, label in enumerate(ax_3.get_xticklabels()):
    if ind % 49 == 0:  # every 10th label is kept
        label.set_visible(True)
    else:
        label.set_visible(False)
ax_3
```

```python
ax_4 = df.groupby('Whiskey_ID').count()['User_ID'].hist(bins = 200);
ax_4.set_xlim(1,50);
ax_4.set_xticks(range(1,150));
ax_4.set_xlabel('Number of Reviews')
ax_4.set_ylabel("Count of user_ID's")
ax_4.grid(False)

for ind, label in enumerate(ax_4.get_xticklabels()):
    if ind % 49 == 0:  # every 10th label is kept
        label.set_visible(True)
    else:
        label.set_visible(False)

ax_4
```
